# Read the dataset to the python environment.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
car_data=pd.read_csv('cars_data.csv')
car_data
```

# Check for the null values present in the dataset.

```python
car_data.isnull().any()
```

# Plot a bar graph of male vs female buyers participated in the sales.

```python
df=sns.barplot(x="Buyer Gender",y='Sale Price',data=car_data)
plt.title( "Gender Vs Sales ",fontsize=16) 
for container in df.containers:
    df.bar_label(container)
```

# Find the top 5 cars based on their sales price.

```python
print('Top 5 cars based on sales are')
top5=car_data.nlargest(5, ['Sale Price'])
top5[['Make','Model', 'Sale Price']]
```

# Find the least 5 cars based on their Resell price.

```python
print('Least 5 cars based on sales are')
Least5=car_data.nsmallest(5, ['Sale Price'])
Least5[['Make','Model', 'Sale Price']]
```
